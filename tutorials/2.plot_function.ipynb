# plot函数
plot是pyplot的画线函数,接收成对的x, y数值，以及用来描述所画图形的颜色，线型和标志的关键字参数。
x, y数值可以为列表，元组或者numpy数组（实际上pyplot总是把列表或元组转换为numpy数组）。其中x值为可选参数，如果没有没有提供第一个参数 x，则plot函数默认使用y数组的索引([0,1,2,3...])作为 x 坐标。
例子：

```python
import numpy as np
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [1,4,9,16]
plot(x,y)
plot(y)
plot(x,y,color='red',linestyle='dashed', marker='o',linewidth=.5,label='line 1')
```

此外，plot函数还可以接收字符参数，可以用字符来控制线条属性。
表示颜色的字符参数有：

| character | color   |
|-----------|---------|
| 'b'       | blue    |
| 'g'       | green   |
| 'r'       | red     |
| 'c'       | cyan    |
| 'm'       | magenta |
| 'y'       | yellow  |
| 'k'       | black   |
| 'w'       | white   |

表示类型的字符参数有：

| character | description           | character | description         |

| --        | dashed line style     | 4         | tri_right marker    |
| -.        | dash-dot line style   | s         | square marker       |
| :         | dotted line style     | p         | pentagon marker     |
| .         | point marker          | *         | star marker         |

| v         | triangle_down marker  | +         | plus marker         |
| ^         | triangle_up marker    | x         | x marker            |
| <         | triangle_left marker  | D         | diamond marker      |
| >         | triangle_right marker | d         | thin_diamond marker |
| 1         | tri_down marker       | _         | hline marker        |
| 2         | tri_up marker         |

例如我们要画出红色圆点：

```python
import numpy as np
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x, y, 'ro')
plt.show()
```

黑色线条：

```python
plt.plot(x, y, 'k-')
plt.show()
```

# #show函数
默认情况下，matplotlib.pyplot 不会直接显示图像，只有调用 show() 函数时，图像才会显示出来。
plt.show() 默认是在新窗口打开一幅图像，并且提供了对图像进行操作的按钮。

多组数据
plot函数可以同时用多组数据画图。


```python
x = np.arange(0., 10., 0.2)
plt.plot(x, x+10, 'k-', 
         x, x**2, 'rs', 
         x, x*2, 'g^')
plt.show()
```

或者使用如下形式：

```python
x = np.arange(0., 10., 0.2)
plt.plot(x, x, 'r--',)
plt.plot(x, x**2, 'bs')
plt.plot(x, x**3, 'g^')
plt.show()
```

设置线条属性
除了使用plot函数直接设定线条属性外，还可以使用plot返回的对象来设置属性。
plot 函数返回一个 Line2D 对象组成的列表，每个对象代表输入的一对组合，例如：
line1, line2 = plt.plot(x1, y1, x2, y2)
line1, line2 为两个 Line2D 对象
而
lines = plt.plot(x1, y1, x2, y2, x3, y3)
返回 3 个 Line2D 对象组成的列表

可以使用setp函数修改线条性质
lines = plt.plot(x, y)
plt.setp(lines, color='r', linewidth=2.0)
lines可以设定的属性如下：

|性质|描述|数据类型|
|------|-------|------|
|alpha|透明度|0~1之间的浮点数|
|animated|动画|True 或 False|
|antialiased or aa|抗锯齿|True 或 False|
|color or c|颜色|matplotlib颜色|
|label|标记|字符串|
|linestyle or ls|线型|['solid' ， 'dashed', 'dashdot', 'dotted'|
|linewidth or lw|线条宽度|浮点数|
|marker|数据点标记|数据点标记类型|
|markersizeor ms|数据点标记大小|浮点数|

完整类型请查看：
http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot



## 载入数据
当数据以文本形式存储时，可以使用numpy的loadtxt 函数来进行读取。
用法：
loadtxt(fname, dtype=<type 'float'>, 
        comments='#', delimiter=None, 
        converters=None, skiprows=0, 
        usecols=None, unpack=False, ndmin=0)

loadtxt 有很多可选参数，其中 
delimiter:用于指定分隔符，默认为空格, eg., delimiter=',' #逗号分割
skiprows:参数表示忽略开头的行数，可以用来读写含有标题的文本, eg., skiprows=1 #忽略第一行, 注意不是从0开始。
dtype:指数据转换为numpy数组的数据类型，eg., dtype=np.float
usecols:指定使用哪几列数据,eg., usecols=(0,1,2,4)
comments:注释符,eg., comments='%' #使用百分号作为注释符
对于逗号分隔的文件（通常为.csv格式），只需要制定分割符的参数即可

% test

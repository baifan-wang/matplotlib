matplotlib提供了多种方式来对图像进行定制。

一、使用参数字典(rcParams)
matplotlib的默认参数存储在matplotlibrc(matplotlib resource configurations)当中，其中规定了视图窗口大小以及dpi，线条宽度，颜色和样式，坐标轴，坐标和网格，文本，字体等属性等。
matplotlib从下面的3个地方按顺序查找matplotlibrc文件：
1. 当前目录下
2. 用户默认的自定义在.matplotlib/matplotlibrc里。
3. 在linux下存放在：/usr/lib/python3.5/site-packages/matplotlib/mpl-data/matplotlibrc。若在windows：C:\pytho3.5\lib\site-packages。每次重新安装matplotlib时，该文件会被覆盖，注意备份。

可以用以下方式查看当前使用的matplotlibrc文件是哪个：

```python
import matplotlib
matplotlib.matplotlib_fname()
```

除了通过文本编辑器来修改matplotlibrc之外，还可以matplotlib.rcParams命令来修改参数。如：

matplotlib.rcParams['xtick.direction'] = 'out'  #设置x轴刻度朝向图像外侧

matplotlib.rcParams['ytick.direction'] = 'out'
matplotlib.rcParams['font.family'] = 'sans-serif'  #设置字体家族

matplotlib.rcParams['font.sans-serif'] = 'Arial'   #设置Arial字体

二、使用内置方法
matplotlib提供了一系列的方法来对图像的各种属性进行设定。

限定坐标轴范围
使用axis方法：
plt.axis([xmin, xmax, ymin, ymax])

或者使用xlim和ylim方法来分别设定x和y轴范围：
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

为x,y轴的主刻度和次刻度设置颜色、大小、方向，以及标签大小。

plt.xticks([0,1,2,3,4])
plt.yticks([0,1,2,3,4],fontsize = 14, color = 'blue', rotation = 45)  #设置y轴刻度标签字体大小为14，蓝色，并旋转45°

给x,y轴添加标签
plt.xlabel('x label')
plt.ylabel('y label',fontsize = 14, color = 'blue', rotation = 45) #设置y轴标签字体大小，颜色和旋转角度

给图像添加标题
plt.title('title')

添加文字说明
plt.text()

text()可以在图中的任意位置添加文字，并支持LaTex语法
text函数的关键词见：
http://matplotlib.org/api/text_api.html#matplotlib.text.Text

添加图例
plt.legend()

legend函数的关键词见：
http://matplotlib.org/api/legend_api.html#matplotlib.legend.Legend.contains

其中规定了图例的位置，大小，边框，字体大小，间距等。
plt.legend(loc='upper left',prop = {'size':10},labelspacing=.2, markerscale=2, handletextpad=.2)

其中 loc 参数可以取 0-10 或者字符串，表示图例放置的位置：

|loc string|loc code|
|---------|---------|
|'best'|0|
|'upper right'|1|
|'upper left'|2|
|'lower left'|3|
|'lower right'|4|
|'right'|5|
|'center left'|6|
|'center right'|7|
|'lower center'|8|
|'upper center'|9|
|'center'|10|

文本注释
plt.annotate()

在数据可视化的过程中，图片中的文字经常被用来注释图中的一些特征。使用annotate()方法可以很方便地添加此类注释。在使用annotate时，要考虑两个点的坐标：被注释的地方xy(x, y)和插入文本的地方xytext(x, y)。

上述的文字标记均支持LaTex语法。在字符串中使用一对 $$ 符号可以利用LaTex语法来输入特殊符号以及数学表达式。通常给字符串加上r标记表明它是一个原始字符串。
比如要输入希腊字母，给上述的函数添加诸如r'$\alpha$', r'$\beta$ + $\gamma$'的字符串即可在图中显示希腊字母：

```python
import matplotlib.pyplot as plt
plt.title(r'$\alpha$'+r'$\beta$ + $\gamma$',fontsize=18)
plt.show()
```

物理单位如度，埃：

```python
plt.title('37 $^\circ$' +'  100 $\AA$',fontsize=18)
plt.show()
```

在此以画图实例来说明对图像的各种修饰

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
x = np.linspace(0, 2 * np.pi)
y1 = np.sin(x)
y2 = np.cos(x) 
matplotlib.rcParams['xtick.direction'] = 'out'   #设置x轴标签往外
plt.plot(x,y1,color='blue',linestyle='-', linewidth=1.,label='sin') 
plt.plot(x,y2,color='red',linestyle='--', marker='s',linewidth=2.,label='cos')
plt.xlim(0,6)  #限制x轴范围为0~6
plt.ylim(-1.2,1.2) #限制x轴范围为-1.2~1.2
plt.xticks([0, np.pi/2, np.pi,3*np.pi/2,2*np.pi],['$0$', '$\pi/2$', '$\pi$','3$\pi/2$','2$\pi$'], fontsize =12)
#设置x轴刻度为0,pi/2, pi,3pi/2,2pi, 同时传入一组文本标签
plt.title('sin and cos plot', fontsize = 18, loc='right') #设置图片标题
plt.xlabel('x axis', fontsize = 14, color = 'blue') #设置x轴标题
plt.ylabel('y axis', fontsize = 14, color = 'red') #设置y轴标题
plt.text(-0.9,1.2,'A',fontsize=15)   #在坐标(-0.9,1.2)出添加一行文字（A）
plt.legend(prop = {'size':10},labelspacing=.4, markerscale=1, handletextpad=.5,loc='upper center')
#设置图例，放在图片上方正中位置
i = np.pi  ##标注途中特定位置，这里标注坐标为pi，sin(pi)
plt.plot(i, np.sin(i),linestyle = '',marker='s',color ='blue') #使用plot函数画一个点,linestyle为空即不显示线条
plt.annotate(r'annotate here', 
             xy=(i, np.sin(i)), # 数据点坐标位置
             xytext=(+5, +20), # 文本位置坐标
             textcoords='offset points', # 坐标相对于数据点的坐标
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.3")) # 画一个箭头
plt.show()
```

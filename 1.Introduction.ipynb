# matplotlib简介

matplotlib 是python最著名的绘图库，提供了丰富的数据绘图工具。
matplotlib的pyplot子模块包括了许多numpy和pyplot中常用的函数，提供了和matlab类似的绘图API，方便用户快速绘制图表。

# 使用pylot
导入相关的模块：

```python
import matplotlib.pyplot as plt
import numpy as np
```

绘制直线图

```python
x = [1,2,3,4]
y = [1,2,3,4]
plt.plot(x,y, 'r-')
plt.show()
```

绘制散点图

```python
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x,y)
plt.show()
```

绘制直方图

```python
x =  np.random.randn(10000)
n, bins, patches = plt.hist(x, 25, facecolor='r')
plt.show()
```

绘制等高线图

```python
data = np.loadtxt('D:\\test.dat', skiprows=0)
X=data[:,0].reshape(73,73)
Y=data[:,1].reshape(73,73)
Z=data[:,2].reshape(73,73)
cp = plt.contourf(X,Y,Z,cmap='RdYlGn')
colorbar = plt.colorbar(cp)
plt.show()
```

绘制饼图

```python
sizes = [10, 40, 20, 30]
colors = ['yellow', 'green', 'navy', 'purple']
explode = (0, 0.1, 0, 0)
plt.pie(sizes, explode = explode, colors = colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

画两个或多个图

```python
x1 = np.linspace(0.0, 4.0)
x2 = np.linspace(0.0, 4.0)
y1 = np.sin(2 * np.pi * x1)
y2 = np.tan(2 * np.pi * x2)
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'go-')
plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'yd-')
plt.show()
```
